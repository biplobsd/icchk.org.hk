```
import re, urllib.request, urllib.parse
from openpyxl import Workbook
# install chromium, its driver, and selenium
from IPython.display import HTML, clear_output
!apt update
!apt install chromium-chromedriver
!pip install selenium
# set options to be headless, ..
clear_output()
from selenium import webdriver
options = webdriver.ChromeOptions()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
# open it, go to a website, and get results
wd = webdriver.Chrome('chromedriver',options=options)

workbook = Workbook()
sheet = workbook.active

# cell head names
sheet["A1"] = "Page URL"
sheet["B1"] = "Company"
sheet["C1"] = "Company Name"
sheet["D1"] = "Address"
sheet["E1"] = "Contact"
sheet["F1"] = "Phone"
sheet["G1"] = "Fax"
sheet["H1"] = "Email"
sheet["I1"] = "Member Type"
sheet["J1"] = "Telex"
sheet["K1"] = "WebSite"
sheet["L1"] = "Subsidary"
sheet["M1"] = "Management"
sheet["N1"] = "Business Nature"
sheet["O1"] = "Import Product"
sheet["P1"] = "Export Product"
sheet["Q1"] = "Sub Product"

totalPage = 38
baseUrl = 'http://www.icchk.org.hk/'
comPage = 2
for page in range(1, totalPage):
  print(page)
  pageurl = urllib.request.urlopen(f"{baseUrl}business_directory.php?page={page}&companyname=&membertype=&char=").read().decode()
  eatchUrl = re.findall(r'"(companydetails.php[^"]+)', pageurl)
  # print(eatchUrl)
  for url in eatchUrl:
    goUrl = urllib.parse.urljoin(baseUrl, url)
    goUrl = urllib.parse.quote(goUrl, safe="/?:&=()")
    print(goUrl)
    sheet[f"A{comPage}"] = goUrl
    
    # without selenium
    htmldata = urllib.request.urlopen(goUrl).read().decode()
    sheet[f"B{comPage}"] = baseUrl+re.search(r'Company</B>.+\n.+src="([^"]+)', htmldata).group(1)
    # sheet[f"C{comPage}"] = re.search(r'Company Name</B>.+\n.+">([^&]+)', htmldata).group(1)
    # sheet[f"D{comPage}"] = re.search(r'Address</B>.+\n.+">([^&]+)', htmldata).group(1)
    # sheet[f"E{comPage}"] = re.search(r'Contact</B>.+\n.+<TD >([^&]+)', htmldata).group(1)
    # sheet[f"F{comPage}"] = re.search(r'Phone</B>.+\n.+<TD>([^&]+)', htmldata).group(1)
    # sheet[f"G{comPage}"] = re.search(r'Fax</B>.+\n.+<TD >([^&]+)', htmldata).group(1)
    # sheet[f"H{comPage}"] = re.search(r'Email</B>.+\n.+<TD>([^&]+)', htmldata).group(1)
    # sheet[f"I{comPage}"] = re.search(r'Member Type</B>.+\n.+<TD>([^&]+)', htmldata).group(1)
    # sheet[f"J{comPage}"] = re.search(r'Telex</B>.+\n.+<TD>([^&]+)', htmldata).group(1)
    # sheet[f"K{comPage}"] = re.search(r'WebSite</B>.+\n.+<TD>([^&]+)', htmldata).group(1)
    # sheet[f"L{comPage}"] = re.search(r'Subsidary</B>.+\n.+<TD>([^&]+)', htmldata).group(1)
    # sheet[f"M{comPage}"] = re.search(r'Management</B>.+\n.+<TD>([^&]+)', htmldata).group(1)
    
    # with selenium
    wd.get(goUrl)
    sheet[f"C{comPage}"] = wd.find_element_by_xpath('//b[contains(., "Company Name")]/../../td[2]').text
    sheet[f"D{comPage}"] = wd.find_element_by_xpath('//b[contains(., "Address")]/../../td[2]').text
    sheet[f"E{comPage}"] = wd.find_element_by_xpath('//b[contains(., "Contact")]/../../td[2]').text
    sheet[f"F{comPage}"] = wd.find_element_by_xpath('//b[contains(., "Phone")]/../../td[2]').text
    sheet[f"G{comPage}"] = wd.find_element_by_xpath('//b[contains(., "Fax")]/../../td[2]').text
    sheet[f"H{comPage}"] = wd.find_element_by_xpath('//b[contains(., "Email")]/../../td[2]').text
    sheet[f"I{comPage}"] = wd.find_element_by_xpath('//b[contains(., "Member Type")]/../../td[2]').text
    sheet[f"J{comPage}"] = wd.find_element_by_xpath('//b[contains(., "Telex")]/../../td[2]').text
    sheet[f"K{comPage}"] = wd.find_element_by_xpath('//b[contains(., "WebSite")]/../../td[2]').text
    sheet[f"L{comPage}"] = wd.find_element_by_xpath('//b[contains(., "Subsidary")]/../../td[2]').text
    sheet[f"M{comPage}"] = wd.find_element_by_xpath('//b[contains(., "Management")]/../../td[2]').text
    sheet[f"N{comPage}"] = wd.find_element_by_xpath('//b[contains(., "Business Nature")]/../../td[2]').text
    sheet[f"O{comPage}"] = wd.find_element_by_xpath('//b[contains(., "Import Product")]/../../td[2]').text
    sheet[f"P{comPage}"] = wd.find_element_by_xpath('//b[contains(., "Export Product")]/../../td[2]').text
    sheet[f"Q{comPage}"] = wd.find_element_by_xpath('//b[contains(., "Sub Product")]/../../td[2]').text
    
    comPage += 1

workbook.save(filename="doneOutput.xlsx")
```
